```python
import pandas as pd
import numpy as np
```

```python
rev_ba = pd.read_csv('./data/BeerAdvocate/ratings-without-text.csv', nrows=8393032)
```

```python
df_ba = pd.read_csv('./data/BA_groups.csv')
```

```python
# Add year to reviews (corresponds to timestamp 'data')
rev_ba['year'] = pd.to_datetime(rev_ba['date'], unit='s').dt.year
rev_ba['year'] = rev_ba['year'].astype(np.int64)

big_ba = df_ba[["year","brewery_name", "main_group"]]
ba_big_ratings = rev_ba.merge(big_ba, how='left', on=["brewery_name", "year"]) 
```

```python
# Sanity check
ba_big_ratings[ba_big_ratings['brewery_name']=='Brauerei Diebels GmbH & Co KG'].head(3)
```

```python
ba_big_ratings.to_csv('./data/BeerAdvocate/ratings_big.csv', index=False)
```
